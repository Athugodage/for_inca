```python
import fasttext
import re

model = fasttext.load_model('lid.176.ftz')

def lab2lang(label):
    comma = str(label).index(',')
    return str(label)[comma-3:comma-1]
    

df['country (language)'] = df['App review text'].apply(lambda x: lab2lang(model.predict(x, k=1)))

```
